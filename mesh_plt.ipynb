```python
import numpy as np
import matplotlib.pyplot as plt
import os


def plot_mesh(mesh_dir):
    X = np.loadtxt(os.path.join(mesh_dir, 'x.dat'))   # (ny+1) × (nx+1)
    Y = np.loadtxt(os.path.join(mesh_dir, 'y.dat'))

    ny, nx = X.shape[0] - 1, X.shape[1] - 1

    fig, ax = plt.subplots(figsize=(7, 7))

    # 画横线（沿 x 方向，每行节点连线）
    for i in range(X.shape[0]):
        ax.plot(X[i, :], Y[i, :], color='steelblue', linewidth=0.5)

    # 画竖线（沿 y 方向，每列节点连线）
    for j in range(X.shape[1]):
        ax.plot(X[:, j], Y[:, j], color='steelblue', linewidth=0.5)

    ax.set_title(f'Mesh  ({ny} × {nx} cells)', fontsize=13)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_aspect('equal', adjustable='box')
    ax.invert_yaxis()          # y 向下为正，与云图一致
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    mesh_dir = input("请输入网格文件夹路径（如 cavity_mesh）: ").strip()
    plot_mesh(mesh_dir)
```
